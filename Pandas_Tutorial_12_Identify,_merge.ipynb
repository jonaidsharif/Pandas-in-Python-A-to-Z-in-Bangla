```python
author_df.merge(book_df) # authod_df is the data and book_df is the same data showing
```

```python
author_df.merge(book_df, how = 'inner', on = 'author_id') # only merged records where both keys matcha
```

```python
author_df.merge(book_df, how = 'outer', on = 'author_id')
# records from both DataFrames are included, even if some keys dn't match
```

```python
author_df.merge(book_df, how = 'left', on = 'author_id')
# all records from the left DataFrame are included, and only matching records from the right DataFrame are included
```

```python
author_df.merge(book_df, how = 'right', on = 'author_id')
# all records from the right DataFrame are included, and only matching records from the left DataFrame are included
```
