```python
import re

def price_convert(row):
  thousand = 1000
  lakh = 100000

  if 'Thousand' in row:
    return float(re.sub(r"[\s]|Thousand|thousand", '', row)) * thousand

  return float(re.sub(r"[\s]lakh", "", row)) * lakh

def area_convert(row):
  return int(re.sub(r"[\s]sqft", "", row))
```

```python
df['price'] = df['price'].apply(price_convert)
df['area'] = df['area'].apply(area_convert)

df['beds'] = df['beds'].str.replace['Bed', ''].astype['int']
df['bath'] = df['bath'].str.replace['Bath', ''].astype['int']

df['city'] = df['adress'].apply(lambda x: 'Dhaka' if 'Dhaka' in x else 'Chattogram')
```
