```python
df = pd.DataFrame(more_people)

df
```

```python
df.loc[4] = ['C101188', 'Farhana', 'Nabi', 'farhananabiiii@gmail.com', 'nascania'] # loc is the index 4 value data changing
```

```python
df.loc[5, 'last_name'] = 'Jonaid' # indexing 5 values, last name which Jibon replace to the Jonaid name
```

```python
df # checking previous data to the updated!!!
```

```python
df.loc[1, ['last_name', 'email']] = 'Ahamd', 'ahmad@gmail.com' # indexing 1 values, last name, email which Ahmad, ahmad@gmail.com replace to the Jonaid name
```

```python
df # checking previous data to the updated!!!
```

```python
df['email'].str.upper() # lower case already here, but temporary upper case updated!!!
```

```python
df['email'] = df['email'].str.upper() # lower case already here, but permanently upper case updated!!!
```

```python
df # checking previous data to the updated!!!
```

```python
df.columns # checking all columns name showing
```

```python
df.columns = ['Matrix ID', 'First_Name', 'Last_Name', 'Email', 'Organization'] # Column rename to the updated!!
```

```python
df.columns # checking all columns name showing
```

```python
df # checking previous data to the updated!!!
```

```python
df.rename(columns = {"First_Name": "Fn", "Last_Name": "Ln"})
# Rename First_Name -> Fn, Last_Name -> Ln changing but its temporary change, not permanently change.
```

```python
df # checking previous data to the updated!!!
```

```python
df.rename(columns = {"First_Name": "Fn", "Last_Name": "Ln"}, inplace = True)
# Rename First_Name -> Fn, Last_Name -> Ln changing but its permanently change
```

```python
df # checking previous data to the updated!!!
```

```python
df.columns = [x.upper() for x in df.columns] # x axis columns name all are the upper case showing
```

```python
df # checking previous data to the updated!!!
```
